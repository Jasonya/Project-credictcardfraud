```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import gridspec 
```

```python
# read the dataset and print out
c_data = pd.read_csv('creditcard.csv')
c_data.describe()
```

```python
# plot out each features distribution
f = c_data.iloc[:, 0:28].columns

plt.figure(figsize=(20,28*4))
gs = gridspec.GridSpec(14, 2)
for i, c in enumerate(c_data[f]):
    ax = plt.subplot(gs[i])
    sns.distplot(c_data[c][c_data.Class == 1], bins=50, color='red')
    sns.distplot(c_data[c][c_data.Class == 0], bins=50, color='green')
    ax.set_xlabel('')
    ax.set_title(f'Historgram of feature: {c}')
plt.show();
```

```python
Fraud = c_data[c_data['Class'] == 1] 
Valid = c_data[c_data['Class'] == 0]
fraud_rate = len(Fraud)/len(Valid)*100
print(f'The fraud_rate: {fraud_rate}')
print(f'fraud transactions: {len(Fraud)}')
print(f'Valid transactions: {len(Valid)}')
```

```python
print('Amount Detials of Fraudulent Transcation')
print(Fraud.Amount.describe())
print('')
print('Amount Detials of Valid Transcation')
print(Valid.Amount.describe())
```

```python
# correlation matrix
cor = c_data.corr()
fig = fig = plt.figure(figsize=(12,9))
# observe no correlation over
sns.heatmap(cor, vmax=0.8, square=True)
plt.show()

# most of the feature is not posstive relative, but some of them are
# negative relative
```

```python
# dividing the X and Y from the dataset
X = c_data.drop(['Class'], axis=1)
Y = c_data['Class']
X = X.values
Y = Y.values
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_Test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
accl, precl, recall, F1l, Matl, tl = [], [], [], [], [], []
```

# IsolationForest

```python
# Isolation Forest Model (normally used in anomaly detection)
import time
from sklearn.ensemble import IsolationForest
st = time.time()
ifc = IsolationForest(max_samples=len(X_train), contamination=fraud_rate, random_state=1)
ifc.fit(X_train)
scores_pred = ifc.decision_function(X_train)
y_pred = ifc.predict(X_test)
et = time.time()
ifc_t = et - st
ifc_t
```

```python
# Reshape the prediction values to 0 for valid, 1 for fraud. 
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1
n_errors = (y_pred != Y_Test).sum()
y_pred
```

```python
#printing the confusion matrix
from sklearn.metrics import confusion_matrix
LABELS = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(Y_Test, y_pred)
plt.figure(figsize=(12, 12))
sns.heatmap(conf_matrix, xticklabels=LABELS,
 yticklabels=LABELS, annot=True, fmt='d');
plt.title('Confusion matrix')
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
#evaluation of the model
#printing every score of the classifier
#scoring in any thing
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef
n_outliers = len(Fraud)
print('the Model used is {}'.format('Isolation Forest'))
acc= accuracy_score(Y_Test,y_pred)
print('The accuracy is {}'.format(acc))
prec= precision_score(Y_Test,y_pred)
print('The precision is {}'.format(prec))
rec= recall_score(Y_Test,y_pred)
print('The recall is {}'.format(rec))
f1= f1_score(Y_Test,y_pred)
print('The F1-Score is {}'.format(f1))
MCC=matthews_corrcoef(Y_Test,y_pred)
print('The Matthews correlation coefficient is{}'.format(MCC))
accl.append(acc)
precl.append(prec)
recall.append(rec)
F1l.append(f1)
Matl.append(MCC)
tl.append(ifc_t)
```

# MLP

```python
from sklearn.neural_network import MLPClassifier
st = time.time()
train_scores, test_scores, generlization_error = [],[], []
# demo - findout the better structure for MLP
for i in range(100,200):
    mlpc = MLPClassifier(hidden_layer_sizes=(i,))
    mlpc.fit(X_train, Y_train)
    y_pred = mlpc.predict(X_test)
    tr_score = mlpc.score(X_train, Y_train)
    te_score = mlpc.score(X_test, Y_Test)
    train_scores.append(tr_score)
    test_scores.append(te_score)
    generlization_error.append(tr_score-te_score)
```

```python
# plot the figure to see the structure change impact
ne = [i for i in range(100,200)]
plt.subplots(figsize=(15,5))
plt.subplots_adjust(wspace=0.4)
plt.subplot(131)
plt.plot(ne, train_scores)
plt.title('Training Score: More complex is better')
plt.xlabel('Model Complexity')
plt.ylabel('Training Score')
plt.subplot(132)
plt.plot(ne, test_scores)
plt.title("Testing Score: There's a sweetspot")
plt.xlabel('Model Complexity')
plt.ylabel('Testing Score')
plt.subplot(133)
plt.plot(ne, np.subtract(train_scores,test_scores))
plt.title("Generalization Error")
plt.xlabel('Model Complexity')
plt.ylabel('(Training Score) - (Testing Score)')
plt.show()
```

```python
best = generlization_error.index(max(generlization_error))
print(f'The best structure: {best}')
```
